```python
def cal_size(num_params,dtype):
    if dtype == "float32":
        size = (num_params/1024**2) * 4    
    elif dtype == "float16":
        size = (num_params/1024**2) * 2
    elif dtype == "int8":
        size = (num_params/1024**2) * 1
    else:
        size = -1
    print("Size of model is : {:.2f} MB".format(size))
    return size
```

```python
cal_size(sum(ok), 'float32')
```
